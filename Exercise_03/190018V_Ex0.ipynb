```python
# imports
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

def imshowBGR(img,ax=plt):
    ax.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
```

```python
# 1

im=cv.imread('butterfly.jpg',cv.IMREAD_REDUCED_GRAYSCALE_4)
assert im is not None

box_kernal=1./81*np.ones((9,9))
im_avg=cv.filter2D(im,-1,box_kernal)

k_size=9
sigma=4
im_gau=cv.GaussianBlur(im,(k_size,k_size),sigma)

fig,ax=plt.subplots(1,3,figsize=(18,6))

ax[0].imshow(im,cmap='gray',vmin=0,vmax=255)
ax[0].set_title('Original')

ax[1].imshow(im_avg,cmap='gray',vmin=0,vmax=255)
ax[1].set_title('Avg')

ax[2].imshow(im_gau,cmap='gray',vmin=0,vmax=255)
ax[2].set_title('Gau')
```

```python
# 2

f=cv.imread('contact_lens.tif',cv.IMREAD_GRAYSCALE).astype(np.float32)
assert f is not None


sobel_v=np.array([[-1,-2,-1],[0,0,0],[1,2,1]],
dtype=np.float32)

f_x=cv.filter2D(f,-1,sobel_v)

sobel_h=np.array([[-1,0,1],[-2,0,2],[-1,0,1]],
dtype=np.float32)

f_y=cv.filter2D(f,-1,sobel_h)

grad_mag=np.sqrt(f_x**2+f_y**2)

fig,ax=plt.subplots(1,4,figsize=(18,6))

ax[0].imshow(f,cmap='gray',vmin=0,vmax=255)
ax[0].set_title('Original')

ax[1].imshow(f_x,cmap='gray',vmin=-1020,vmax=1020)
ax[1].set_title('f_x')

ax[2].imshow(f_y,cmap='gray',vmin=-1020,vmax=1020)
ax[2].set_title('f_y')

ax[3].imshow(grad_mag,cmap='gray')
ax[3].set_title('grad_mag')

```
